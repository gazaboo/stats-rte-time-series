### <a id='toc1_4_1_'></a>[Solution : Charger les données](#toc0_)

```python
# imports
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import glob
rcParams['figure.figsize'] = 15, 5
```

```python
# Cleaner alternative using pipe (Flo 2023) :
def create_datetime_index(df):
    df = df.copy()
    df.index = pd.to_datetime(df['Date'] + ' ' + df['Heures'])
    df = df.drop(['Date', 'Heures'], axis='columns')
    return df


def load_data(files):
    # Ceci est un générateur (économe en ressources)
    df_national = (
        pd.read_csv(f, encoding='ISO-8859-1', sep='\t', index_col=False, na_values=['ND', '-']) for f in files
    )

    # Utilisation de pipe
    df = (
        pd.concat(df_national)
        .dropna(
            subset=['Date', 'Heures', 'Consommation'],
            how='any')
        .fillna(0)
        .pipe(create_datetime_index)
    )

    return df
```

```python
# Chargement des données nationales
FOLDER = './data_rte/'
files = glob.glob(FOLDER + 'eCO2mix_RTE_Annuel-Definitif' + '*.zip')
df_national = load_data(files)

df_national.head()
```

```python
# Chargement des données régionales
FOLDER = './data_rte/'
files = [FOLDER + f for f in os.listdir(FOLDER)
         if not f.startswith('eCO2mix_RTE_Annuel-Definitif')]
df_regional = load_data(files)

df_regional.head()
```

```python
pivot = pd.pivot_table(
    df_regional,
    values='Consommation',
    index=df_regional.index,
    columns="Périmètre"
)

pivot.stack().reset_index()
```
